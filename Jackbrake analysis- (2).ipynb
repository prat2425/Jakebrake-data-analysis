# Columns with continous data-

```python
continous=[]

for i in df.columns:
    if df[i].dtypes=='int64':
        continous.append(i)
```

```python
continous
```

```python
for i in df.columns:
    if df[i].dtypes=='float64':
        continous.append(i)
```

```python
continous
```

```python
plt.figure(figsize=(30,45))
plotnumber=1

for col in continous:
    if plotnumber<=35:
        ax=plt.subplot(9,3,plotnumber)
        sns.distplot(df[col],color='r')
        plt.xlabel(col,fontsize=16)
        plt.xticks(fontsize=14)
        plt.yticks(rotation=0,fontsize=16)
    plotnumber+=1
plt.show() 
```

# Observations-

1. Here we can see that Hour,minute,miles,Min rate,average rate per mile,total time taken and max rate has the outliers which can be clearly seen in the above diagram as the distribution is right skewed.

```python
df.skew()
```

# Observations-

1. Here we can see that all the columns except minute and seconds are having skewness more than 0.5 which shows they are not following standard normal distribution and consist of outliers.

# Checking outliers with box plot-

```python
plt.figure(figsize=(30,45))
plotnumber=1

for col in continous:
    if plotnumber<=35:
        ax=plt.subplot(9,3,plotnumber)
        sns.boxplot(df[col],color='r')
        plt.xlabel(col,fontsize=16)
        plt.xticks(fontsize=14)
        plt.yticks(rotation=0,fontsize=16)
    plotnumber+=1
plt.show()
```

As mentioned above we can clearly see that all the columns except Minute and seconds are having outliers which can be seen in the above plot.

# Removing outliers-

```python
#upper side outliers-q3+(1.5*IQR)
#lower side outliers-q1-(1.5*IQR)


q1=df.quantile(0.25)
q3=df.quantile(0.75)

IQR=q3-q1
```

```python
q1
```

```python
q3
```

# Removing outliers from max_rate column-

```python
rate_high=(q3.max_rate+(1.5*IQR.max_rate))


index=np.where(df['max_rate']> rate_high)

df=df.drop(df.index[index])
df
```

# Removing outliers from min_rate column-

```python
rate_high=(q3.min_rate+(1.5*IQR.min_rate))


index=np.where(df['min_rate']> rate_high)

df=df.drop(df.index[index])
df
```

# Removing outliers from miles column-

```python
rate_high=(q3.miles+(1.5*IQR.miles))


index=np.where(df['miles']> rate_high)

df=df.drop(df.index[index])
df
```

# Removing outliers from total time taken column-

```python
rate_high=(q3.total_time_taken+(1.5*IQR.total_time_taken))


index=np.where(df['total_time_taken']> rate_high)

df=df.drop(df.index[index])
df
```

So we can say now these are the most genuine loads.

```python
plt.figure(figsize=(30,45))
plotnumber=1

for col in continous:
    if plotnumber<=35:
        ax=plt.subplot(9,3,plotnumber)
        sns.distplot(df[col],color='r')
        plt.xlabel(col,fontsize=16)
        plt.xticks(fontsize=14)
        plt.yticks(rotation=0,fontsize=16)
    plotnumber+=1
plt.show() 
```
